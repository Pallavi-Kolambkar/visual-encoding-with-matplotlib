```python
sizes=dataset['Generation'].value_counts(dropna=True)
labels=sizes.index
```

The pie charts use the encoding of angle,area and arc length. The pie chart helps in visulaizing the dominance of a group in comparison to other groups. The disadvantage of pie charts is that when there are more sub parts the graphs become really difficult to read. Even though the graphs cover the encoding accurately they can't always be used to reprent the information accurately. However to understand the distribution of data and see the individual share of data pie chart can be useful. In this particular graph it is easier to compare the count of each type of generations, thus pointing out the dominance of the generation 1 pokemons.

The bar chart is the easier graph in terms of visualization amongst the two graphs because it uses the length encoding to convey the counts which makes it easier to compare and deduce conclusions. The colors help in seperating the generations and highlighting the largest ones. The only drawback is to measure the exact quantities in which case the pie chart is useful because even though there is a 0.1 difference in the counts of pokemon for generations 1 and 5 reflects the changes accordingly.
